```python
import cv2
import numpy as np
import os
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from joblib import dump, load
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            images.append(img)
    return images

def extract_features(image, method='sift'):
    if method == 'sift':
        keypoints, descriptors = detect_and_compute_sift(image)
    elif method == 'orb':
        keypoints, descriptors = detect_and_compute_orb(image)
    else:
        raise ValueError("Method must be 'sift' or 'orb'")
    return descriptors

def detect_and_compute_sift(image):
    sift = cv2.SIFT_create(contrastThreshold=0.02)
    keypoints, descriptors = sift.detectAndCompute(image, None)
    return keypoints, descriptors

def detect_and_compute_orb(image):
    orb = cv2.ORB_create(nfeatures=2000)
    keypoints, descriptors = orb.detectAndCompute(image, None)
    return keypoints, descriptors

def train_model(images, method='sift'):
    all_descriptors = []

    for image in images:
        descriptors = extract_features(image, method)
        if descriptors is not None:
            all_descriptors.append(descriptors)
    
    all_descriptors = np.vstack(all_descriptors)

    scaler = StandardScaler()
    all_descriptors = scaler.fit_transform(all_descriptors)

    oc_svm = OneClassSVM(kernel='rbf', gamma='auto').fit(all_descriptors)

    return oc_svm, scaler

def predict_logo_with_confidence(model, scaler, image, method='sift'):
    descriptors = extract_features(image, method)
    if descriptors is not None:
        descriptors = scaler.transform(descriptors)
        prediction = model.predict(descriptors)
        confidence_scores = model.decision_function(descriptors)
        return prediction, confidence_scores
    return None, None

def evaluate_model_with_confidence(loaded_model, loaded_scaler, real_images, fake_images, method='sift'):
    y_true = []
    y_pred = []
    y_confidence = []

    # Predict real images (should be predicted as '1')
    for image in real_images:
        prediction, confidence_scores = predict_logo_with_confidence(loaded_model, loaded_scaler, image, method=method)
        if prediction is not None:
            y_true.extend([1] * len(prediction))
            y_pred.extend(prediction)
            y_confidence.extend(confidence_scores)
    
    # Predict fake images (should be predicted as '-1')
    for image in fake_images:
        prediction, confidence_scores = predict_logo_with_confidence(loaded_model, loaded_scaler, image, method=method)
        if prediction is not None:
            y_true.extend([-1] * len(prediction))
            y_pred.extend(prediction)
            y_confidence.extend(confidence_scores)

    accuracy = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=['Fake (-1)', 'Real (1)'])
    conf_matrix = confusion_matrix(y_true, y_pred)

    return accuracy, report, conf_matrix, y_confidence

# Evaluate the model with confidence scores
accuracy, report, conf_matrix, confidence_scores = evaluate_model_with_confidence(loaded_model, loaded_scaler, real_images, fake_images, method=method)

# Print the accuracy, classification report, and confidence scores
print(f"Model Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)

# Optionally, print confidence scores
for i, score in enumerate(confidence_scores):
    print(f"Sample {i + 1}: Confidence Score = {score:.4f}")


def save_model(model, scaler, model_path, scaler_path):
    dump(model, model_path)
    dump(scaler, scaler_path)

def load_model(model_path, scaler_path):
    model = load(model_path)
    scaler = load(scaler_path)
    return model, scaler

def evaluate_model(loaded_model, loaded_scaler, real_images, fake_images, method='sift'):
    y_true = []
    y_pred = []

    # Predict real images (should be predicted as '1')
    for image in real_images:
        prediction = predict_logo(loaded_model, loaded_scaler, image, method=method)
        if prediction is not None:
            y_true.extend([1] * len(prediction))
            y_pred.extend(prediction)
    
    # Predict fake images (should be predicted as '-1')
    for image in fake_images:
        prediction = predict_logo(loaded_model, loaded_scaler, image, method=method)
        if prediction is not None:
            y_true.extend([-1] * len(prediction))
            y_pred.extend(prediction)

    accuracy = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=['Fake (-1)', 'Real (1)'])
    conf_matrix = confusion_matrix(y_true, y_pred)

    return accuracy, report, conf_matrix
```

```python
# Set the path to the real logos folder
real_data_folder_path = 'data/original'  # Replace with the actual path to real logos

# Load real logo images
real_images = load_images_from_folder(real_data_folder_path)

# Train the model with real logos
method = 'sift'  # Choose 'sift' or 'orb'
model, scaler = train_model(real_images, method=method)

# Save the model and scaler
save_model(model, scaler, 'logo_oc_svm_model.joblib', 'logo_scaler.joblib')
```

```python
# Set the path to the fake logos folder
fake_data_folder_path = 'data/fake'  # Replace with the actual path to fake logos

# Load fake logo images
fake_images = load_images_from_folder(fake_data_folder_path)

# Load the saved model and scaler
loaded_model, loaded_scaler = load_model('logo_oc_svm_model.joblib', 'logo_scaler.joblib')

# Evaluate the model
accuracy, report, conf_matrix = evaluate_model(loaded_model, loaded_scaler, real_images, fake_images, method=method)

# Print the accuracy and classification report
print(f"Model Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Fake', 'Real'], yticklabels=['Fake', 'Real'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
